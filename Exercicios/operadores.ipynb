### Exercicio 1: Operadores Aritméticos

Calcula a área e perímetro dun rectángulo:
- Pide ao usuario base e altura.
- Calcula área.
- Calcula perímetro.
- Utiliza `//` e `%`.
- Amosa os resultados utilizando diferentes formatos de cadeas.

```python
base= int(input("Introduce la base del rectángulo: "))
altura= int(input("Introduce la altura del rectángulo: "))

area= base * altura

perimetro= (base*2) + (altura*2)

proba_operador_1= base//altura
proba_operador_2= base%altura

print(f"F1. La area es {area} y el perimetro es {perimetro}")
print("F2. La area es {} y el perimetro es {}".format(area,perimetro))
print("F3. La area es ",area, " y el perimetro es ",perimetro)
```

### Exercicio 4: Operadores Lóxicos

Determina se un ano é bisesto:
- Divisible entre 4.
- Non divisible entre 100, excepto se é divisible entre 400.
- Non encadenes condicionais senón condicións.
- Exemplo: 2024 (bisesto), 1900 (non bisesto), 2000 (bisesto).

```python

ano = int(input("Introduce un ano para saber se é bisesto: "))

e_bisesto = (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0)

if e_bisesto:
    print(f"O ano {ano} é bisesto")
else:
    print(f"O ano {ano} non é bisiesto")
```

### Exercicio 9: Combinación de Operadores

Calculadora de IMC (xa a fixemos pero desta volta é máis completa):
- peso / altura²
- Clasifica segundo táboa OMS
- Baixo peso: < 18.5
- Normal: 18.5 - 24.9
- Sobrepeso: 25 - 29.9
- Obesidade: >= 30
- Usa operadores de comparación e lóxicos
- Opcional: walrus para evitar repetir cálculo

```python
peso = float(input("Introduce o teu peso (kg): "))
altura = float(input("Introduce a túa altura (m): "))

print(f"\nO teu IMC é: {(imc := round(peso / (altura ** 2), 2))}")


if imc < 18.5:
    clasificacion = "Baixo peso"
elif 18.5 <= imc <= 24.9: 
    clasificacion = "Peso normal"
elif imc >= 25 and imc <= 29.9: 
    clasificacion = "Sobrepeso"
else: 
    clasificacion = "Obesidade"

print(f"Estado segundo a OMS: **{clasificacion}**")
```

### Exercicio 10: Analizador de Texto

Analiza un texto introducido polo usuario:
1. Lonxitude.
2. Se contén certas palabras.
3. Conta vogais.
4. Verifica se é palíndromo.
5. Verifica se texto é `None`.
6. Concatena.

```python
texto = input("Introduce unha frase para analizar: ")

e_nulo = texto is None


print(f"\nAnalizando... Lonxitude: {(n := len(texto))} caracteres.")

contén_python = "python" in texto.lower()
contén_ola = "ola" in texto.lower()


total_vogais = sum(1 for caracter in texto.lower() if caracter in "aeiouáéíóú")

texto_limpo = texto.replace(" ", "").lower()
e_palindromo = texto_limpo == texto_limpo[::-1]


resultado_final = "Análise final: " + texto.upper()

print("-" * 30)
print(f"¿É a entrada nula (is None)?: {e_nulo}")
print(f"¿Contén a palabra 'Python'?: {contén_python}")
print(f"Número total de vogais: {total_vogais}")
print(f"¿É un palíndromo (léese igual ao revés)?: {e_palindromo}")
print(f"Slicing inverso (reverso): {texto[::-1]}")
print(resultado_final)
```

### Xogo de Adiviñanza

Xogo onde:
- Número secreto aleatorio (1-100).
- Aportar Pistas ("é maior", "é menor").
- Contador de intentos.
- Utiliza Walrus.
- Preguntar se se quere xogar outra vez.

```python
import random


xogar_de_novo = "si"

while xogar_de_novo.lower() == "si":
    numero_secreto = random.randint(1, 100)
    intentos = 0
    print("\nAcabo de pensar un número entre 1 e 100. Podes adiviñalo?")

    while (intento := int(input(f"Intento {intentos + 1}: "))) != numero_secreto:
        intentos += 1
        
        if intento < numero_secreto:
            print("Pista: O número secreto é MAIOR.")
        else:
            print("Pista: O número secreto é MENOR.")
            
    intentos += 1 
    print(f"\n¡PARABÉNS! Adiviñaches o número {numero_secreto} en {intentos} intentos.")
    
    xogar_de_novo = input("\nQueres xogar outra vez? (si/non): ")

print("\nGrazas por xogar. Ata a próxima!")
```

### Validador de Email

Valida email con:
- Que conteña unha `@`.
- Que conteña un `.` despois da `@`.
- Lonxitude >= 5.
- Non conteña espazos en branco.
- Dominio pertenza a unha lista de dominios válida.

```python
dominios_validos = ["gmail.com", "outlook.com", "edu.xunta.gal", "python.org"]


email = input("Introduce o teu correo electrónico: ").strip()

long_ok = len(email) >= 5

ten_arroba = "@" in email

sen_espazos = " " not in email

if ten_arroba:
    parte_dominio = email.split("@")[-1]
    ten_punto_dominio = "." in parte_dominio
else:
    parte_dominio = ""
    ten_punto_dominio = False

dominio_autorizado = parte_dominio in dominios_validos


e_valido = long_ok and ten_arroba and sen_espazos and ten_punto_dominio and dominio_autorizado

print(f"Email analizado: {email}")
print(f"¿Ten @?: {ten_arroba}")
print(f"¿Sen espazos?: {sen_espazos}")
print(f"¿Punto no dominio?: {ten_punto_dominio}")
print(f"¿Dominio '{parte_dominio}' autorizado?: {dominio_autorizado}")

print("-" * 30)
if (resultado := e_valido):
    print("O email é VÁLIDO.")
else:
    print("O email NON é válido. Revisa os requisitos.")
```
