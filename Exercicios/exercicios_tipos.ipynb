A partires dos seguintes elementos:

2. numero_float a enteiro (qué pasa coa parte decimal?)

```python
numero_str = "1234"
numero_float = 45.67
lista_numeros = [1, 2, 3, 4, 5]

num_int = int(numero_str)

float_a_int = int(numero_float)

tupla_num = tuple(lista_numeros)
conxunto_num = set(lista_numeros)

suma_str = str(int(numero_str) + numero_float)

print(f"1. Enteiro: {num_int} ({type(num_int)})")
print(f"2. Float a Int: {float_a_int} (O .67 desapareceu)")
print(f"3. Tupla: {tupla_num} | Conxunto: {conxunto_num}")
print(f"4. String combinado: '{suma_str}'")
```

### Ejercicio 5: Tuplas

Crea unha tupla coas coordenadas (x, y, z)

```python
coordenadas = (3, 7, 2)
```

Realiza estas operacións:

1. Accede a cada coordenada individualmente
2. Intenta modificar unha coordenada (qué ocorre?)
3. Convirte a tupla a lista
4. Crea unha nova tupla concatenando (1, 2, 3)
5. Desempaqueta a tupla en variables x, y, z

```python
coordenadas = (3, 7, 2)

x_coord = coordenadas[0]
y_coord = coordenadas[1]
z_coord = coordenadas[2]
print(f"1. Acceso individual: x={x_coord}, y={y_coord}, z={z_coord}")

try:
    coordenadas[0] = 10
except TypeError as e:
    print(f"2. Erro ao modificar: {e}")

lista_coords = list(coordenadas)
lista_coords[0] = 10  
print(f"3. Lista convertida: {lista_coords}")

nova_tupla = coordenadas + (1, 2, 3)
print(f"4. Tupla concatenada: {nova_tupla}")

x, y, z = coordenadas
print(f"5. Variables desempaquetadas: x={x}, y={y}, z={z}")
```

### Exercicio 6: Diccionarios

A partires do seguinte diccionario:

```python
estudante = {
    "nome": "Carlos",
    "idade": 35,
    "modulos": ["Redes", "Sistemas", "Programación"],
    "activo": True
}
```

Realiza estas operacións:

1. Engade a chave "promedio" con valor 8.5
2. Actualiza la idade a 40
3. Engade "Hardware" á lista de módulos
4. Amosa todas as chaves do diccionario
5. Crea unha copia do diccionario e modifica a copia

```python
estudante = {
    "nome": "Carlos",
    "idade": 35,
    "modulos": ["Redes", "Sistemas", "Programación"],
    "activo": True
}

estudante.update({"promedio":8.5,})

estudante["idade"]=40

estudante["modulos"].append("Hardware")

print(estudante.keys())

copia_estudante= estudante.copy()

copia_estudante.update({"altura":1.80})

print(copia_estudante)
```

Crea unha lista de diccionarios con información de estudantes:

Realiza estas operacións:

### Exercicio 8: Conxuntos

A partir destes conxuntos:

```python
conx_a = {1, 2, 3, 4, 5}
conx_b = {4, 5, 6, 7, 8}
```

Realiza estas operacións:

1. Unión de ambos conxuntos
2. Intersección
3. Diferencia (A - B)
4. Diferencia simétrica
5. Verifica si 3 está en A

```python
conx_a = {1, 2, 3, 4, 5}
conx_b = {4, 5, 6, 7, 8}


union = conx_a | conx_b 
print(f"1. Unión: {union}")


interseccion = conx_a & conx_b  
print(f"2. Intersección: {interseccion}")


diferenza = conx_a - conx_b  
print(f"3. Diferenza (A - B): {diferenza}")


dif_simetrica = conx_a ^ conx_b  
print(f"4. Diferenza simétrica: {dif_simetrica}")


esta_presente = 3 in conx_a
print(f"5. ¿Está o 3 en conx_a?: {esta_presente}")
```

### Exercicio 9: Programa Integrador

Crea un programa que:

1. Pida ao usuario o seu nome, idade e cidade de residencia
2. Almacene esa información nun diccionario
3. Convirta a idade a enteiro
4. Cree unha lista de hobbies (pregunta por 3 polo menos)
5. Amose toda a información formateada de xeito lexible

```python
nome= input("Introduce nombre")
idade= int(input("Introduce idade"))
cidade_residencia=input("Introduce a cidade de residencia")

hobbies=[]

añadir_hobbie=True

while añadir_hobbie or len(hobbies)<3:
    hobbie=input("Introduce un hobbie")
    hobbies.append(hobbie)
    if len(hobbies)>=3:
        añadir_hobbie=input("Quieres añadir un hobbie más? si/no").lower()
        if añadir_hobbie!= "si":
            añadir_hobbie=False

persona={"nome":nome, "idade":idade, "cidade de reidencia":cidade_residencia,"hobbies":hobbies}

for clave,valor in persona.items():
    print(f"{clave}:{valor}")
    print("="*20)

```

### Exercicio 10: Combinación de Tipos Complexos

Temos un Sistema de biblioteca:

```python
biblioteca = {
    "libros": [
        {
            "titulo": "Cien años de soledad",
            "autor": "Gabriel García Márquez", 
            "ano": 1967,
            "xeneros": {"Realismo mágico", "Novela"},
            "disponible": True
        },
        {
            "titulo": "1984",
            "autor": "George Orwell",
            "ano": 1949,
            "xeneros": {"Ciencia ficción", "Distopía"},
            "disponible": False
        }
    ],
    "usuarios": {
        "001": {"nome": "Ana", "libros_prestados": ["1984"]},
        "002": {"nome": "Carlos", "libros_prestados": []}
    }
}
```

Realiza estas operacións:

1. Engade un novo libro.
2. Marca un libro como non dispoñible.
3. Atopa todos os libros dun xénero.
4. Libros dispoñibles por autor.
5. Usuario con máis libros prestados.

```python
biblioteca = {
    "libros": [
        {
            "titulo": "Cien años de soledad",
            "autor": "Gabriel García Márquez", 
            "ano": 1967,
            "xeneros": {"Realismo mágico", "Novela"},
            "disponible": True
        },
        {
            "titulo": "1984",
            "autor": "George Orwell",
            "ano": 1949,
            "xeneros": {"Ciencia ficción", "Distopía"},
            "disponible": False
        }
    ],
    "usuarios": {
        "001": {"nome": "Ana", "libros_prestados": ["1984"]},
        "002": {"nome": "Carlos", "libros_prestados": []}
    }
}

novo_libro = {
    "titulo": "Follas Novas",
    "autor": "Rosalía de Castro",
    "ano": 1880,
    "xeneros": {"Poesía", "Rexurdimento"},
    "disponible": True
}
biblioteca["libros"].append(novo_libro)

for libro in biblioteca["libros"]:
    if libro["titulo"] == "Cien años de soledad":
        libro["disponible"] = False


xenero_buscado = "Ciencia ficción"
libros_xenero = [l["titulo"] for l in biblioteca["libros"] if xenero_buscado in l["xeneros"]]

print(f"Libros de {xenero_buscado}: {libros_xenero}")        


autor_buscado = "Rosalía de Castro"
disponibles = [l["titulo"] for l in biblioteca["libros"] if l["autor"] == autor_buscado and l["disponible"]]

print(f"Dispoñibles de {autor_buscado}: {disponibles}")


usuario_top = max(biblioteca["usuarios"].values(), key=lambda u: len(u["libros_prestados"]))

print(f"O usuario con máis préstamos é: {usuario_top['nome']}")
```
